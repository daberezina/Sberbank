```python
%%time

cols = [3,5,6,7,8,9]

for i in flat[flat.full_sq <= flat.life_sq].index[:1]:
    a = flat.loc[i].values
    s = []
    for j in flat.index:
        b = flat.loc[j].values
        if sum((a[cols] - b[cols]) * a[cols] * b[cols]) == 0:
            s.append(b[2])
    if len(s) > 0:
        flat.loc[i] = list(a[:2]) + [np.median(s)] + list(a[3:])
```
